Se descargan los outputs directamente en la web: "https://www.ebi.ac.uk/Tools/services/rest/ncbiblast/result/{job_id}/out", siendo job_ID el anterior

```python
# Definir el directorio donde están los resultados
directory = Path('/Users/Sara/Documents/TFM/Código para entregar/PeptidosPSCA')

# Inicializar una lista para almacenar los mejores candidatos
Candidatos = []

# Iterar sobre cada archivo en el directorio que termine en '.txt'
for file_path in directory.glob('*.txt'):
    with file_path.open('r') as file:
        data = file.readlines()

    # Procesar cada línea para extraer la información deseada
    results = []
    for line in data:
        if line.startswith('SP:'):
            parts = line.split()
            identifier = parts[0]
            Interactor = file_path.name.replace('peptide_', '')
            Interactor = Interactor.replace('_results.txt', '')
            RMSDscore = round(rmsd[int(float(Interactor) - 1)],2)
            score = float(parts[-2])
            e_value = float(parts[-1])
            results.append({'Identifier': identifier, 'Score': score, 'E-value': e_value, 'Interactor' : Interactor, 'RMSD' : RMSDscore})

    # Convertir a DataFrame
    if results:
        df = pd.DataFrame(results)
        
        # Seleccionar el mejor candidato (puedes ajustar este criterio)
        Mejores2 = df.nlargest(2, 'Score')
        Candidatos.extend(Mejores2.to_dict('records'))

# Convertir la lista de los mejores candidatos a un DataFrame para manejo más fácil
Candidatosdf = pd.DataFrame(Candidatos)
Candidatosdf.head()
```

```python
# Definir una función para extraer el nombre 
def Proteinas_con_combre (Tabla_IDs):
    n=0
    for ID in Tabla_IDs['Identifier']:
        ID = ID.replace('SP:', '')
        ID = ID.split("-")
        ID = ID[0]
        # Conectarse a la base de datos de SwissProt y guardar la información asociada a un ID de la proteina
        handle = ExPASy.get_sprot_raw(ID)
        # Guardar la información que lee de lo que ha buscado antes
        record = SwissProt.read(handle)
        # Asignar el nombre de la proteina y añadirlo a la tabla en una nueva columna "Nombre_Proteina"
        Proteina = record.entry_name
        Tabla_IDs.at[n,"Nombre_Proteina"] = Proteina
        Tabla_IDs.at[n,"ID"] = ID
        # Busca en las referencias extraidas de la búsqueda que empiece con HGNC (HUGO Gene Nomenclature Committee),
        # si empieza por eso, el nombre se encuentra en el tercer elemento, lo guarda en una nueva columna de la tabla
        for reference in record.cross_references:
            if reference[0].startswith('HGNC'):
                Gen = reference[2]
                Tabla_IDs.at[n,"Nombre_Gen"] = Gen
                break
        n = n+1
    return(Tabla_IDs)

CandidatosNombre = Proteinas_con_combre (Candidatosdf)
CandidatosNombre
```

```python
CandidatosSig = CandidatosNombre[CandidatosNombre['E-value'] <= 0.05]
CandidatosSig = CandidatosSig.reset_index()
CandidatosSig
```

```python
CandidatosRMSD = CandidatosSig.sort_values('RMSD', ascending=True)
CandidatosRMSD
```

```python
CandidatosRMSD = CandidatosRMSD.drop_duplicates(subset=['ID', 'Nombre_Gen'])
CandidatosRMSD
```
